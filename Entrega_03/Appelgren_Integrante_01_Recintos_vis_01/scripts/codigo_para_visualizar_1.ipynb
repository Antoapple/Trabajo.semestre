```python
import pandas as pd
import altair as alt
data = pd.read_csv('/content/recintos_actualización_tabla_dinámica.csv', delimiter=';')
data
```

```python
data.columns = data.columns.str.strip()

data['Capacidad máxima'] = pd.to_numeric(data['Capacidad máxima'], errors='coerce')

data = data.dropna(subset=['Capacidad máxima'])
```

```python
totales = data['PROPOSITO'].value_counts().to_dict()
```

```python
data['PROPOSITO con total'] = data['PROPOSITO'].map(
    lambda x: f"{x} ({totales.get(x, 0)})"
)
```

```python
chart = alt.Chart(data).mark_bar(opacity=0.7).encode(
    x=alt.X('Capacidad máxima:Q',
            bin=alt.Bin(maxbins=20, step=5000),
            title='Capacidad máxima'),
    y=alt.Y('count()', stack='zero', title='Cantidad de recintos'),
    color=alt.Color('PROPOSITO con total:N', title='Propósito (Total)',scale=alt.Scale(range=['#cf0638','#0a6789','#fecd23','#fa6632','#0a996f']))
).properties(
    width=500,
    height=300
)

chart
```
